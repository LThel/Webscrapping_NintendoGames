```python
def grab_total_shipped(p):
    url = f"https://www.vgchartz.com/games/games.php?page={p}&console=3DS&order=Sales&ownership=Both&direction=DESC&showtotalsales=1&shownasales=0&showpalsales=0&showjapansales=0&showpublisher=1&showdeveloper=0&showreleasedate=1&showlastupdate=0&showvgchartzscore=1&showcriticscore=1&showuserscore=1"
    html_doc = pd.read_html(url)    
    #Grab the columns on the page 
    df = html_doc[-2]
    if p == 1:
        dfA = pd.DataFrame()
        dfB = pd.DataFrame()
        dfA['Names'] = df.iloc[:,2]
        dfA['Total_shipped'] = df.iloc[:,8]
    elif p == 24 :
        #dfB = pd.DataFrame()
        dfB['Names'] = df.iloc[:,2]
        dfB['Total_shipped'] = df.iloc[:,8]        
        dfA = pd.concat([dfA, dfB])
        
    elif p > 1 :
        dfB['Names'] = df.iloc[:,2]
        dfB['Total_shipped'] = df.iloc[:,8]        
        dfA = pd.concat([dfA, dfB])
    return dfA  

#x = pd.DataFrame(map(grab_total_shipped, range(1,4)))
#print(x)
grab_total_shipped(1)
```

```python
#Webscrapping Algorithm : Select the platform and the last page before running the algorithm - Need to automatize
platform = 'N64' # iOS:4 ; NS:45 ; 3DS:24 ; DS:66 ; DSi:2 ; Wii:34 ; WiiU:12 ; GBA:34 ; GBC:1 ; N64:8
last_page = 8
for p in range(1,last_page+1):
    url = f"https://www.vgchartz.com/games/games.php?page={p}&console={platform}&order=Sales&ownership=Both&direction=DESC&showtotalsales=1&shownasales=0&showpalsales=0&showjapansales=0&showothersales=0&showpublisher=1&showdeveloper=0&showreleasedate=1&showlastupdate=0&showvgchartzscore=1&showcriticscore=1&showuserscore=1&showshipped=1"
    html_doc = pd.read_html(url)    
    #Grab the columns on the page 
    df = html_doc[-2]
    if p == 1:
        dfA = pd.DataFrame()
        dfB = pd.DataFrame()
        dfA['Names'] = df.iloc[:,2]
        dfA['Total_shipped'] = df.iloc[:,8]
        dfA['Total_sales'] = df.iloc[:,9]
    elif p == last_page :
        dfB = pd.DataFrame()
        dfB['Names'] = df.iloc[:,2]
        dfB['Total_shipped'] = df.iloc[:,8]
        dfB['Total_sales'] = df.iloc[:,9]
        dfA = pd.concat([dfA, dfB])
        if platform == '3DS':
            dfA.to_csv(f"df_TroisDS.csv")
        else:
            dfA.to_csv(f"df_{platform}.csv")
    elif p > 1 :
        dfB['Names'] = df.iloc[:,2]
        dfB['Total_shipped'] = df.iloc[:,8]
        dfB['Total_sales'] = df.iloc[:,9]
        dfA = pd.concat([dfA, dfB])
```

```python
# Cleaning the resulting df
df_concat.isna().sum()/df_concat.shape[0] #We have more values in Total_sales than in Total_shipped

#We can decide to select either only the games with a Total_sales values 
#or the games that has at least one of the 2 values. 
#But to do so, we have to find a way to compare Total_sales and Total_shipped. I didn't find a smart way to do so

#1st option : df_concat["Names"][(df_concat['Total_shipped'].isna()) & (df_concat['Total_sales'].isna())]
df_concat = df_concat[['Names', 'Total_sales']][df_concat['Total_sales'].isna()==False]
df_concat[df_concat['Total_sales']=='0.00m'].shape[0] #Too see how many games with 0.00m we have 
#=>We can wonder if this is a real information because it is very unlikey to have 443 games with 0.00 m of sales

#We could find a way to link the titles of this dataframe to our other dataframe from kaggle by doing some text analysis
df_concat.rename(columns = {'Names':'Titles','Total_sales':'Total sales in million'}, inplace = True)#Rename our columns
df_concat["Total sales in million"]=df_concat['Total sales in million'].str.replace('m', '').astype(float)
df_concat.reset_index(inplace = True)
df_concat.drop(['index'], axis = 1, inplace = True)
```
